```python
def edit_house(match):
    """
    Reduce data for Representatives/Senators
    
    INPUT
    match the regex serach matches, retrieve with match.group(i)
    
    OUTPUT
    rep, 1 LastName 0
         2 FirstName 1 
         3 Party 2
         4 City 3
         5 Position 4
         6 District 5
         7 
         8 MadisonOffice 6
         9 Telephone 7
         10 Fax 8
         11 
         12 DistrictPhone 9
         13 Email 10
         14 
         15 VotingAddress 11
         16 
         17 Staff 12
        
    
    """
    
    rep = []
    
    index = [1,2,3,4,5,6,8,9,10,12,13,15,17]
    for i in index:
        try:
            tmp = match.group(i)
        except IndexError:
            rep.append(None)
            continue
        
        if not tmp:
            rep.append(None)
            continue
        
        if i == 8 or i == 15:
            # Room 113 NorthState CapitolP.O. Box 8952Madison, WI 53708
            # 10 Division St.Milton, WI 53563
            tmp = re.sub(r'([a-z]|\d|\.)([A-Z])', r'\1 \2', tmp)

        elif i == 9:
            # (608) 266-3790(888) 534-0043
            tmp = re.sub(r'(\d)\(', r'\1, (', tmp)
        elif i == 17 and tmp:
            tmp = filter(None, tmp.rstrip().split('\n'))
            tmp = [(tmp[j], tmp[j+1]) for j in xrange(0, len(tmp), 2)]
        
        rep.append(tmp)
    
    return rep
```

```python
def get_house(out = 'assembly'):
    """
    Get all data for Representatives/Senators
    
    INPUT
    out the legislative house
    
    SAVE TO FILE out.csv
    data
    
    SAVE TO FILE out_list.txt
    names of all the representatives
    
    """
    
    house = []
    
    # get html
    url = 'https://docs.legis.wisconsin.gov/2015/legislators/' + out
    text = load_txt(url)
    
    # parse html
    parser = BeautifulSoup(text, "lxml")
    info = parser.body.find_all('div', attrs={'class':'rounded'})
    
    # set up regex for data
    # https://regex101.com/r/3JOnrh/1
    regex = r'\n(.+?), (.+?) \(([R|D]) - (.+?)\)\s+(.+|)\s+District (\d+?)\n\n' + \
            r'Details\s+(Website|)\s+Madison Office:\n\s+(.+)\s+Telephone:\n\s+(.+)\s+' + \
            r'Fax:\n\s+(.+)\s+(District Phone:\n(.+)|)\s+Email:\n(.+)\s+' + \
            r'(Voting Address:\n\s+(.+)|)\s+(Staff:\n((.|\n)+)|)'
    regexp = re.compile(regex)
    
    # set up regex for each personal website
    ow = r'/2015/legislators/' + out + r'/'
    pw = r'http://legis.wisconsin.gov/' + out + '/'
    official_web = re.compile(ow)
    personal_web = re.compile(pw)
    
    # for each representative
    for inf in info:
        # ignore class="breadcrumb rounded"
        if inf['class'][0] == 'breadcrumb': continue
        
        # match for each representative's data
        match = regexp.search(inf.text)
        
        # check
        if not match:
            warning('No data parsed for representative. HTML below.', inf.text)
        
        # edit data
        rep = edit_house(match)

        # find websites
        # official is your 4digit ID at the end of 
        #     http://docs.legis.wisconsin.gov/2015/legislators/OUT/
        # personal is your District/LastName at the end of
        #     http://legis.wisconsin.gov/OUT/
        # a couple of people don't have personal websites
        official, personal = None, None
        off_head = 'http://docs.legis.wisconsin.gov'
        for link in inf.find_all('a'):
            tmp = link.get('href')
            if tmp:
                if official and personal: break
                tmp = link.get('href')
                if official_web.match(tmp) is not None:
                    official = off_head + tmp #tmp[-4:]
                if personal_web.match(tmp):
                    personal = tmp
        
        # go to official site and grab committees
        committees = None
        if official: 
            ####
            ####
            ####
            #### hmm maybe should go direct to his personal page
            pass
        else: 
            warn = 'Website missing: official'
            warning(warn, official, inf.find_all('a'))
        
        # add data 
        rep.extend((official, personal, committees))
        
        # save each representative
        house.append(rep)
        
    print len(house)
    return house
```
